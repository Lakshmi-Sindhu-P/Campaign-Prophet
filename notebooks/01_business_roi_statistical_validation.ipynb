# Marketing Campaign ROI Optimization & Customer Response Prediction

## Project Overview

### Objective
Analyze a financial marketing campaign to measure campaign effectiveness, estimate return on investment (ROI), validate key business assumptions statistically, and build predictive models to identify customers most likely to subscribe to a term deposit product.

### Business Problem Statement
A financial institution conducts outbound marketing campaigns to promote a term deposit product. Each customer contact incurs a cost, while successful subscriptions generate revenue.

The organization seeks to answer the following strategic questions:
- Which customer segments generate the highest marketing ROI?
- Which customer and campaign factors significantly influence conversion?
- Can we predict which customers are most likely to subscribe?
- How should marketing budgets be allocated to maximize profitability?
- What data-driven recommendations can improve campaign performance?

### ROI Assumptions

The dataset does not contain actual campaign costs, revenue values, deposit amounts, or customer lifetime value. Therefore, ROI is estimated using scenario-based assumptions.

The project uses two ROI frameworks:

1. **Illustrative USD Framework**
   - Conservative Scenario: \$15 cost per contact and \$300 revenue per successful subscription
   - Base Case Scenario: \$10 cost per contact and \$500 revenue per successful subscription
   - Optimistic Scenario: \$5 cost per contact and \$1,000 revenue per successful subscription

2. **Data-Anchored EUR / CLV Framework**
   - Conservative Scenario: €12 cost per contact and €150 estimated CLV per successful subscription
   - Base Case Scenario: €10 cost per contact and €250 estimated CLV per successful subscription
   - Optimistic Scenario: €7 cost per contact and €450 estimated CLV per successful subscription

The EUR framework is more contextually aligned with the Portuguese banking dataset, while the USD framework is retained as an illustrative sensitivity analysis.

### Why ROI Assumptions Are Required

The dataset does not contain actual campaign costs or revenue values.
Business assumptions are introduced to estimate profitability and
compare the relative ROI of customer segments.

### Currency Convention

The dataset originates from a Portuguese retail bank and all
underlying transactions occurred in euros (€). However, because the
dataset contains no actual financial figures — no deposit amounts,
no costs, no revenues — the ROI framework introduces assumed values
rather than observed ones.

The analysis is structured in two stages:

**Stage 1 — USD Reference Framework (Sections 2.2.1 through 2.2.6)**
Financial assumptions are expressed in US dollars ($) as illustrative
reference values. This convention is consistent with academic and
industry practice when using this dataset for methodological
demonstration, where the specific currency is secondary to the
analytical framework being established. All ROI conclusions drawn
from this stage should be interpreted as directional and comparative
rather than as precise financial forecasts.

**Stage 2 — EUR Data-Anchored Framework (Section 2.2.7)**
Financial assumptions are re-expressed in euros (€) and grounded in
publicly available Portuguese banking statistics for the 2008–2010
campaign period, including Banco de Portugal published deposit rate
data and retail banking benchmarks. This stage tests whether the
profitability conclusions from Stage 1 hold under empirically
grounded assumptions in the correct currency context.

The dual-framework structure serves as a robustness check: if segment
rankings and profitability conclusions are consistent across both
stages, confidence in the analytical findings is substantially
strengthened regardless of the specific assumption values used.

### 2.2.1 Define ROI Scenarios
Three business scenarios are defined to estimate campaign profitability under conservative, base-case, and optimistic financial assumptions.

```python
# Define ROI scenarios for sensitivity analysis
roi_scenarios = {
    "Conservative": {
        "cost_per_contact": 15,
        "revenue_per_success": 300
    },
    "Base Case": {
        "cost_per_contact": 10,
        "revenue_per_success": 500
    },
    "Optimistic": {
        "cost_per_contact": 5,
        "revenue_per_success": 1000
    }
}

# Display the scenarios
print("ROI Scenarios Defined:\n")

for scenario, values in roi_scenarios.items():
    print(
        f"{scenario}: "
        f"Cost per Contact = ${values['cost_per_contact']}, "
        f"Revenue per Success = ${values['revenue_per_success']}"
    )
```

#### Inference-4 from ROI Scenario Definition

A multi-scenario financial framework was established to evaluate campaign profitability under varying business assumptions.

**Key Findings**
- Three ROI scenarios were defined: Conservative, Base Case, and Optimistic.
- The Conservative scenario assumes higher marketing costs and lower customer value.
- The Base Case scenario provides a balanced and realistic estimate of profitability.
- The Optimistic scenario assumes lower acquisition costs and higher customer lifetime value.
- All scenarios use the same analytical framework, enabling direct comparison of profitability across customer segments.

**Overall Conclusion**

The scenario-based approach enables sensitivity analysis to determine whether customer segment rankings remain stable under different financial assumptions, improving the robustness of the ROI analysis.

###2.2.2 Calculate ROI by Customer Segment
Campaign performance is aggregated by customer job category, and profitability metrics are computed for each ROI scenario.

```python
# Aggregate campaign performance by job category
job_summary = (
    df.groupby("job")
      .agg(
          total_contacts=("success", "count"),
          successful_contacts=("success", "sum"),
          success_rate=("success", "mean"),
          avg_duration=("duration", "mean")
      )
      .reset_index()
)

# Add estimated direct call cost per contact
# Assumes fully-loaded agent cost of $30/hr (midpoint of $20-$40 range)
job_summary['est_direct_cost'] = (
    job_summary['avg_duration'] / 3600 * 30
).round(2)

print("Job summary with estimated direct call costs:")
display(job_summary[['job', 'total_contacts', 'success_rate',
                       'avg_duration', 'est_direct_cost']].round(4))
```

```python
# Calculate ROI metrics for each scenario
roi_results = {}

for scenario, values in roi_scenarios.items():
    cost_per_contact = values["cost_per_contact"]
    revenue_per_success = values["revenue_per_success"]

    scenario_df = job_summary.copy()

    scenario_df["total_cost"] = (
        scenario_df["total_contacts"] * cost_per_contact
    )

    scenario_df["total_revenue"] = (
        scenario_df["successful_contacts"] * revenue_per_success
    )

    scenario_df["net_profit"] = (
        scenario_df["total_revenue"] - scenario_df["total_cost"]
    )

    scenario_df["roi_percent"] = (
        scenario_df["net_profit"] / scenario_df["total_cost"]
    ) * 100

    # Sort by ROI from highest to lowest
    scenario_df = scenario_df.sort_values(
        "roi_percent",
        ascending=False
    ).reset_index(drop=True)

    roi_results[scenario] = scenario_df
```

#### Objective

#### Analytical Focus
- Apply EV = p × R − C at the segment level using observed conversion rates
- Identify which segments generate positive expected value per contact
- Quantify total expected profit under targeted vs untargeted outreach
- Establish the per-contact profit framework that Phase 5 extends to
  individual customers using XGBoost predicted probabilities

#### Expected Business Outcome
By the end of this section, each job segment will have an expected

```python
# 2.2.6 Expected Value by Job Segment (Base Case)

cost = roi_scenarios['Base Case']['cost_per_contact']
revenue = roi_scenarios['Base Case']['revenue_per_success']
breakeven = breakeven_rates['Base Case']

ev_analysis = job_summary.copy()

# EV = p × R − C
ev_analysis['expected_value_per_contact'] = (
    ev_analysis['success_rate'] * revenue - cost
).round(2)

ev_analysis['contact_decision'] = np.where(
    ev_analysis['success_rate'] > breakeven,
    'CONTACT',
    'DO NOT CONTACT'
)

ev_analysis['total_expected_profit'] = (
    ev_analysis['expected_value_per_contact'] *
    ev_analysis['total_contacts']
).round(0)

ev_analysis = ev_analysis.sort_values(
    'expected_value_per_contact', ascending=False
).reset_index(drop=True)

print("Expected Value by Job Segment — Base Case ($10 cost / $500 revenue)")
print(f"Breakeven conversion rate: {breakeven:.4f} "
      f"({breakeven*100:.2f}%)\n")

display(ev_analysis[[
    'job',
    'total_contacts',
    'success_rate',
    'expected_value_per_contact',
    'total_expected_profit',
    'contact_decision'
]].round(2))

contact_segs = ev_analysis[ev_analysis['contact_decision'] == 'CONTACT']
no_contact_segs = ev_analysis[ev_analysis['contact_decision'] == 'DO NOT CONTACT']

print(f"\nSegments warranting contact     : "
      f"{len(contact_segs)} of {len(ev_analysis)}")
print(f"Segments to deprioritize        : "
      f"{len(no_contact_segs)} of {len(ev_analysis)}")
print(f"\nTotal expected profit — untargeted : "
      f"${ev_analysis['total_expected_profit'].sum():,.0f}")
print(f"Total expected profit — targeted   : "
      f"${contact_segs['total_expected_profit'].sum():,.0f}")
print(f"Profit improvement from targeting  : "
      f"${contact_segs['total_expected_profit'].sum() - ev_analysis['total_expected_profit'].sum():,.0f}")
```

```python
# 3.1 Job Category vs Subscription Success - Chi-Square Test

# Create contingency table
job_subscription_table = pd.crosstab(df['job'], df['y'])

# Perform chi-square test of independence
chi2_job, p_value_job, dof_job, expected_job = stats.chi2_contingency(job_subscription_table)

# Calculate effect size using Cramer's V
n_job = job_subscription_table.values.sum()
cramers_v_job = np.sqrt(
    chi2_job / (n_job * (min(job_subscription_table.shape) - 1))
)

print("3.1 Job Category vs Subscription Success")
print("Chi-Square Test of Independence")
print("-" * 60)
print(f"Chi-square statistic: {chi2_job:.4f}")
print(f"Degrees of freedom: {dof_job}")
print(f"P-value: {p_value_job:.10f}")
print(f"Cramer's V: {cramers_v_job:.4f}")

if p_value_job < 0.05:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: Job category and subscription success are statistically associated.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: No statistically significant relationship was detected.")
```

```python
# 3.1 Job-Level Subscription Summary

job_subscription_summary = (
    df.groupby('job')['y']
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
)

job_subscription_summary['total_customers'] = (
    job_subscription_summary['yes'] + job_subscription_summary['no']
)

job_subscription_summary['subscription_rate_%'] = (
    job_subscription_summary['yes'] / job_subscription_summary['total_customers'] * 100
).round(2)

job_subscription_summary = job_subscription_summary.rename(
    columns={
        'yes': 'subscribers',
        'no': 'non_subscribers'
    }
)

job_subscription_summary = job_subscription_summary[
    ['job', 'total_customers', 'subscribers', 'non_subscribers', 'subscription_rate_%']
].sort_values(
    by='subscription_rate_%',
    ascending=False
)

display(job_subscription_summary)
```

- Cramer's V: 0.1528

## 3.3 Call Duration vs Subscription Success (Independent Sample T-Test)

### Objective

Determine whether successful and unsuccessful subscriptions differ significantly in average call duration.

## Business Questions to Be Addressed

| MQ/CQ ID | Business Question | Status Before Section |
|--------|----------------|----------------|
| MQ11 / CQ2 | Do successful and unsuccessful customers differ in average call duration? | Pending |

### What to Expect in This Section

This section compares the average duration of campaign calls for customers who subscribed versus customers who did not subscribe.

### Analytical Focus

- Separate call duration by subscription outcome
- Calculate group-level summary statistics
- Perform an independent sample t-test
- Interpret statistical significance
- Explain why call duration is explanatory but not available for pre-contact targeting

### Expected Business Outcome

By the end of this section, we will determine whether call duration differs significantly between successful and unsuccessful subscriptions, while also clarifying its operational limitation for proactive campaign targeting.

```python
# 3.3 Call Duration vs Subscription Success - Independent Sample T-Test

# Split call duration into two groups
duration_success = df[df['y'] == 'yes']['duration']
duration_failure = df[df['y'] == 'no']['duration']

# Perform Welch's independent sample t-test
t_duration, p_value_duration = stats.ttest_ind(
    duration_success,
    duration_failure,
    equal_var=False
)

# Create duration summary table
duration_summary = pd.DataFrame({
    'subscription_outcome': ['Subscribed', 'Not Subscribed'],
    'count': [duration_success.count(), duration_failure.count()],
    'mean_duration_seconds': [duration_success.mean(), duration_failure.mean()],
    'median_duration_seconds': [duration_success.median(), duration_failure.median()],
    'std_duration_seconds': [duration_success.std(), duration_failure.std()]
})

duration_summary[['mean_duration_seconds', 'median_duration_seconds', 'std_duration_seconds']] = (
    duration_summary[['mean_duration_seconds', 'median_duration_seconds', 'std_duration_seconds']]
    .round(2)
)

```

```python
print("3.3 Call Duration vs Subscription Success")
print("Independent Sample T-Test")
print("-" * 60)
print(f"T-statistic: {t_duration:.4f}")
print(f"P-value: {p_value_duration:.10f}")
print("Significance level: 0.05")

if p_value_duration < 0.05:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: Average call duration differs significantly between subscribers and non-subscribers.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: No statistically significant difference was detected.")

display(duration_summary)
```

#### Inference-10 from Call Duration vs Subscription Success

The independent sample t-test confirms that successful subscriptions are associated with significantly longer call durations.

**Statistical Results**
- Mean Duration (Successful Subscriptions): 553.26 seconds
- Mean Duration (Unsuccessful Subscriptions): 220.87 seconds
- Median Duration (Successful Subscriptions): 449.00 seconds
- Median Duration (Unsuccessful Subscriptions): 164.00 seconds
- T-Statistic: 55.4984
- P-Value: < 0.0001
- Significance Level: 0.05

**Key Findings**
- Successful subscriptions involve calls that are, on average, more than twice as long as unsuccessful calls.
- The p-value is far below the significance threshold, leading to rejection of the null hypothesis.
- Call duration is strongly associated with customer conversion outcomes.
- The median duration also shows a large difference, with successful calls lasting 449 seconds compared with 164 seconds for unsuccessful calls.
- The magnitude of the difference suggests that engagement level during the conversation may be an important indicator of customer interest.

**Overall Conclusion**

Call duration is strongly associated with subscription success, but this relationship is not causal. Longer calls likely reflect customer interest rather than directly causing conversion — engaged customers stay on the line longer. Because call duration is unknown before outreach begins, it cannot be used for pre-contact targeting. Its value in this analysis is explanatory, not operational.

- Interpret statistical significance
- Distinguish statistical significance from practical business impact

### Expected Business Outcome

By the end of this section, we will determine whether customer age differs significantly between subscribers and non-subscribers, and whether that difference is practically meaningful for campaign targeting.

```python
# 3.4 Age vs Subscription Success - Independent Sample T-Test

# Split age into two groups
age_success = df[df['y'] == 'yes']['age']
age_failure = df[df['y'] == 'no']['age']

# Perform Welch's independent sample t-test
t_age, p_value_age = stats.ttest_ind(
    age_success,
    age_failure,
    equal_var=False
)

# Create age summary table
age_summary = pd.DataFrame({
    'subscription_outcome': ['Subscribed', 'Not Subscribed'],
    'count': [age_success.count(), age_failure.count()],
    'mean_age': [age_success.mean(), age_failure.mean()],
    'median_age': [age_success.median(), age_failure.median()],
    'std_age': [age_success.std(), age_failure.std()]
})

age_summary[['mean_age', 'median_age', 'std_age']] = (
    age_summary[['mean_age', 'median_age', 'std_age']]
    .round(2)
)

print("3.4 Age vs Subscription Success")
print("Independent Sample T-Test")
print("-" * 60)
print(f"T-statistic: {t_age:.4f}")
print(f"P-value: {p_value_age:.10f}")
print("Significance level: 0.05")

if p_value_age < 0.05:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: Average age differs significantly between subscribers and non-subscribers.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: No statistically significant difference was detected.")

display(age_summary)
```
